# Atractor de Lorenz, solucionando las ecuaciones
# dx/dt = a*(y-x)
# dy/dt = x*(b-z)-y
# dz/dt = x*y-c*z

```python
import pandas as pd
import numpy as np
from scipy.integrate import odeint
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
a = 10 
b = 28 #b=28 ~ caos, b = 99.96 nudo t√≥rico
c = 8/3
x0= 1.0
y0= 1.0
z0= 1.0
t = np.linspace(0,100,100000)
```

```python
def deriv(l,t, a, b, c):
    x, y, z = l
    dxdt    = a*(y-x)
    dydt    = x*(b-z)-y
    dzdt    = x*y - c*z
    return dxdt, dydt, dzdt
```

```python
l0 = x0, y0, z0
ret = odeint(deriv, l0, t, args=(a, b, c))
x,y,z = ret.T
```

```python
plt.plot(x,z)
plt.show()
```
